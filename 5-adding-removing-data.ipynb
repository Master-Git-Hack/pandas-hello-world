Si se quiere agregar una columna intermedia entre las columnas 'a' y 'b', se puede hacer con el metodo insert() de pandas.
parametros:
1. la posicion donde se quiere insertar la nueva columna
2. el nombre de la nueva columna
3. la operacion que se quiere hacer para asignar los valores de la nueva columna

```python
df.insert(1, 'nueva_columna', df['a'] - df['b'])
```

````python
df.insert(2,"viewsPerLike",df["views"]/df["likes"])
df.tail()
````

Para remover una columna se puede hacer con el metodo drop() de pandas.
parametros:
1. el nombre de la columna que se quiere eliminar
2. axis=1 para indicar que se quiere eliminar una columna
3. inplace=True para que se modifique el dataframe original

```python
df.drop('nueva_columna', axis=1, inplace=True)
```

````python
df = df.drop("day",axis=1)
df.tail()
````

Para insertar filas en un dataframe se puede hacer con el metodo append() de pandas.
parametros:
1. el dataframe que se quiere agregar
2. ignore_index=True para que se reindexen las filas

```python
df = df.append({'a': 1, 'b': 2}, ignore_index=True)
```

````python
df2 = pd.DataFrame([[1,3,4,7,9],[2,4,5,8,10],[3,5,6,9,11]],columns=["A","B","C","D","E"])
df2
````

````python
df2 = pd.DataFrame([[1,3,4,7,9],[2,4,5,8,10],[3,5,6,9,11]],columns=df.columns)
df2
````

append() was deprecated in pandas version 1.4.0 and was removed in version 2.0.0. To fix this, you should use the pd.concat() function, which is the recommended method for appending DataFrames
parametros:
1. el dataframe que se quiere agregar
2. ignore_index=True para que se reindexen las filas

```python
df = pd.concat([df, pd.DataFrame({'a': [1], 'b': [2]})], ignore_index=True)
```

````python

df = pd.concat([df, df2], ignore_index=True)
df.tail()
````
