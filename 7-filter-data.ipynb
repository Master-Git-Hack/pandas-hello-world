para implementar una condicion anidada, no es igual que con python, se debe de hacer con & y |, en lugar de and y or, esto porque lo que hace es comparar los valores de los arreglos y no los arreglos en si, bit por bit

```python
def array_diff(a, b):
    return [x for x in a if x not in b]
```
