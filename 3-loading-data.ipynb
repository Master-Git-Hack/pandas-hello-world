en la columna date el valor es un string, pero esta en formato de fecha, por lo que se puede convertir a un formato de fecha con la funcion pd.to_datetime

```python
df["day"]
# para acceder a una columna, esto se hace de forma muy similar a un diccionario, y se puede acceder a una columna por su nombre. retorna toda la columna
```

```python
df["day"]=pd.to_datetime(df["day"],format="%Y-%m-%d")
```
