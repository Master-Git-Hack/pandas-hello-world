```python
df_clean = df.dropna(how='all', axis=1)  # Eliminar columnas vacías
df=df_clean
df_clean
```

```python
# Combinar las columnas Unnamed: 1 a Unnamed: 5 en una nueva columna llamada 'Account'
#merge columns unnamed 1 to 5 into a single one, keeping the first non-null value
df_clean['Account'] = df_clean[['Unnamed: 1', 'Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4', 'Unnamed: 5']].fillna('').agg(' '.join, axis=1)

# Eliminar las columnas anteriores
df_clean = df_clean.drop(columns=['Unnamed: 1', 'Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4', 'Unnamed: 5'])

# Mover la columna 'Account' a la primera posición
cols = ['Account'] + [col for col in df_clean.columns if col != 'Account']
df_clean = df_clean[cols]


```
