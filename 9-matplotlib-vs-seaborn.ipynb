```python
%matplotlib inline
%load_ext fireducks.pandas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("./datasets/live-by-subscribed-2021.csv",index_col=0)
df["day"]=pd.to_datetime(df["day"],format="%Y-%m-%d")
df.head()
```

```python
fig = plt.figure()
plt.boxplot(
    [
        (df["views"])[df["subscribedStatus"]=="SUBSCRIBED"],
        (df["views"])[df["subscribedStatus"]=="UNSUBSCRIBED"]
    ]
)
plt.xlabel("subscribedStatus")
plt.ylabel("views")
#ticks = set(df["subscribedStatus"])# it is random order
ticks = ("SUBSCRIBED","UNSUBSCRIBED")
plt.xticks(ticks=range(1,len(ticks)+1),labels=ticks)
fig.show()
```

con matplotlib todo es manual, te da un control preciso de lo que quieres realizar, pero a veces es tedioso. En este caso, se puede hacer un gráfico de barras con una sola línea de código.

```python
fig = plt.figure()
sns.boxplot(
    x="subscribedStatus",
    y="views",
    data=df
)
fig.show()
```
