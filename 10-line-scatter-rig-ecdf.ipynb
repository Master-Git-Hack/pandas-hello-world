```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("./datasets/second-half-2020.csv", index_col=0)
df["day"] = pd.to_datetime(df["day"], format="%Y-%m-%d")
df.head()
```

to_datetime parametros:
- year: int
- month: int
- day: int
- hour: int
- minute: int
- second: int
- microsecond: int
- nanosecond: int
- tz: Union[str, datetime.tzinfo, None] = None
- format: Union[str, None] = None
- errors: str = 'raise'
- infer_datetime_format: bool = False (se refiere a si se debe inferir el formato de la cadena de fecha y hora)
- dayfirst: bool = False
- yearfirst: bool = False
- utc: bool = None
- box: bool = True (si es True, devuelve un DataFrame si se pasa una lista de datos)
- format: str = None
- exact: bool = True
- unit: str = None (si se pasa, se convierte a la unidad especificada, si no, se convierte a datetime64[ns])
- origin: Union[str, datetime.datetime, Timestamp] = 'unix' (puede ser 'unix
- cache: bool = True
tipo de errores:
- 'raise': si hay errores, se lanza un ValueError
- 'coerce': si hay errores, se establece NaT (Not a Time)
- 'ignore': si hay errores, se devuelve la entrada original
- 'warn': si hay errores, se imprime un aviso y se devuelve la entrada original
- 'default': si hay errores, se establece NaT (Not a Time)
- 'backward': si hay errores, se establece NaT (Not a Time)
- 'forward': si hay errores, se establece NaT (Not a Time)
- 'quietly': si hay errores, se establece NaT (Not a Time)


lineplot de sns recibe de parametros:
- x: eje x
- y: eje y
- data: dataframe
- hue: color de la linea
- style: estilo de la linea
- markers: marcadores
- dashes: tipo de linea
- ci: intervalo de confianza
- estimator: estimador
- ax: eje
- legend: leyenda
- palette: paleta de colores
- hue_order: orden de los colores
- style_order: orden de los estilos
- units: unidades
- estimator: estimador
- n_boot: numero de bootstraps(bootstrap es una tecnica de remuestreo)
- sort: ordenar
- err_style: estilo de errorhue="country"
- err_kws: parametros del error
- legend: leyenda

```python
fig = plt.figure()
sns.lineplot(data=df, x="views", y="day",)
fig.show()
```

```python
fig = plt.figure()
sns.scatterplot(data=df, x="views", y="estimatedMinutesWatched", hue=df["dislikes"] > 0)
fig.show()
```

```python
fig = plt.figure()
sns.scatterplot(data=df, x="views", y="estimatedMinutesWatched", hue=df["dislikes"] > 0)
sns.rugplot(data=df, y="views", hue=df["dislikes"] > 0, alpha=0.5)
fig.show()
```

```python
fig = plt.figure()
sns.ecdfplot(data=df, x="views", stat="count")
fig.show()
```
