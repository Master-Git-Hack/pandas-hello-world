```python
%matplotlib inline
%load_ext fireducks.pandas
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("./datasets/revenue-2021.csv", index_col=0)
df["day"] = pd.to_datetime(df["day"], format="%Y-%m-%d")
df.head()
```

```python
df2=df[["estimatedRevenue","estimatedAdRevenue","grossRevenue"]]
df2.head()
```

```python
fig = plt.figure()
sns.kdeplot(data=df,x="grossRevenue")
fig.show()
```

```python
fig = plt.figure()
sns.kdeplot(data=df2,fill=True,cumulative=True)
fig.show()
```

```python
fig = plt.figure()
sns.histplot(data=df2,x="estimatedAdRevenue",bins=20)
fig.show()
```

```python
fig = plt.figure()
sns.histplot(data=df2,element="poly",alpha=0.3,kde=True)
fig.show()
```

```python
fig = plt.figure()
sns.histplot(data=df,x="views",y="cpm",cbar=True,cbar_kws={"shrink":0.75})
#sns.scatterplot(data=df,x="views",y="cpm",color="red")
fig.show()
```
